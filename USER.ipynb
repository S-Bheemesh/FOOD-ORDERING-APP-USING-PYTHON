```python
import csv
from prettytable import PrettyTable

class user:
    def __init__(self):
        pass      
    
    def run(self):
        print("enter R for Registration ")
        print("enter L for login ")
        self.ans=input()
        if self.ans.upper()=="R":
            self.Registration()
        elif self.ans.upper()=="L":
            self.login()
    
    def Registration(self):
        self.f_name=input("Enter your full name :")
        self.p_num=input("Enter your phone number :")
        self.email=input("Enter your Email :")
        self.address=input("Enter your address :")
        self.pas=input("Set your password :")
        with open("USERS.csv","r") as users:
            data=csv.reader(users)
            for row in data:
                if row[2]==self.email:
                    print("You are already a user try logging in")
                    self.login()
                    return
        with open("USERS.csv","a",newline="") as user:
            write=csv.writer(user)
            write.writerow([self.f_name,self.p_num,self.email,self.address,self.pas])
            
        print("\t'NOW YOU ARE A USER'")
        print("\t'LOGIN TO ORDER'")
        self.login()              
            
        
    def login(self):
        self.mail=input("enter your Email :")
        self.paas=input("enter your password :")
        
        with open("USERS.csv","r") as user:
            data=csv.reader(user)
            for row in data:
                if self.mail==row[2] and self.paas==row[-1]:
                    print("\n\t'ENTER P TO PLACE NEW ORDER'")
                    print("\t'ENTER O FOR ORDER HISTORY")
                    print("\t'ENTER U TO UPDATE PROFILE'")
                    self.user_ans=input()
                    
                    if self.user_ans.upper()=="P":
                        print("---------------------------MENU---------------------------")
                        tbl=PrettyTable(["food_id","name","Quantity","Price","Discount","Stock"])
                        with open("MENU.csv","r") as menu:
                            data=csv.reader(menu)
                            for row in data:
                                tbl.add_row(row)
                                tbl.horizontal_char="-"
                                tbl.verticle_char="|"
                                tbl.junction_char="+"
                            print(tbl)
                        print("enter food_id of items you want to order :")
                        self.order=input().split()
                        with open("MENU.csv","r") as menu:
                            data=csv.reader(menu)
                            for row in data:                            
                                for i in self.order:                                
                                    if row[0]==i:
                                        with open("ORDER_HISTORY.csv","a",newline="") as user:
                                            write=csv.writer(user)
                                            if row[-1]==self.mail:
                                                pass
                                            else:
                                                row.append(self.mail)
                                            write.writerow(row)
                        with open("MENU.csv","r") as menu:
                            data=csv.reader(menu)
                            orderd=PrettyTable(["food_id","name","Quantity","Price","Discount","Stock"])
                            for row in data:
                                for i in self.order:
                                    if i==row[0]:
                                        orderd.add_row(row)
                                        orderd.horizontal_char="-"
                                        orderd.verticle_char="|"
                                        orderd.junction_char="+"
                            print("YOU HAVE ORDERD : ")
                            print(orderd)                        

                    elif self.user_ans.upper()=="O":
                        with open("ORDER_HISTORY.csv","r") as order:
                            data=csv.reader(order)
                            oh=PrettyTable(["food_id","name","Quantity","Price","Discount","Stock"])

                            for row in data:
                                if row[-1]==self.mail:
                                    oh.add_row(row[:-1])
                                    oh.horizontal_char="-"
                                    oh.verticle_char="|"
                                    oh.junction_char="+"
                            print("Your ORDER_HISTORY : ")
                            print(oh)
                           
                            
                    elif self.user_ans.upper()=="U":
                        with open('USERS.csv', 'r') as csvfile:
                            reader = csv.reader(csvfile)
                            rows = list(reader)
                            
                        for i,row in enumerate(rows):
                            if row[2]==self.mail and row[-1]==self.paas:
                                self.f_name=input("Enter your new full name :")
                                self.p_num=input("Enter your new phone number :")
                                self.email=input("Enter your new Email :")
                                self.address=input("Enter your new address :")
                                self.pas=input("Set your new password :")  
                                new_values=[self.f_name,self.p_num,self.email,self.address,self.pas]
                                rows[i]=new_values
                                
                        with open('USERS.csv', 'w', newline='') as csvfile:
                            writer = csv.writer(csvfile)
                            writer.writerows(rows)
                            print("PROFILE UPDATED SUCCESSFULLY")
                else:
                    print("Enter correct details")
                            
obj=user()
```
